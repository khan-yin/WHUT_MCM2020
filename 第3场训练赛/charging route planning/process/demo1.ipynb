```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(3,5)
def fr(x):
    return 10.4*6470/(3600-9284/x)

plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False 
plt.plot(x,fr(x))

plt.ylabel("传感器的容量(mA/h)")
plt.xlabel("充电的速率r(J/s)")
plt.grid()
plt.savefig('fr.png')
plt.show()
```

```python
x=np.linspace(10,24)
def fu(x):
    return 10.4*25880/(14400-546*x)

plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False 
plt.plot(x,fu(x))

plt.ylabel("传感器的容量(mA/h)")
plt.xlabel("充电器电压(V)")
plt.grid()
plt.savefig('fu.png')
plt.show()
```

```python
x=np.linspace(3,5)
def fv(x):
    return (14000+45920/(x))/492

plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False 
plt.plot(x,fv(x))

plt.ylabel("传感器的容量(mA/h)")
plt.xlabel("充电器移动速率(m/s)")
plt.grid()
plt.savefig('fv.png')
plt.show()

```

```python
x=np.linspace(1,5)
def ft(x):
    return (14400*x+11420)/492

plt.rcParams['font.sans-serif'] = ['SimHei'] 
# plt.rcParams['axes.utnicode_minus'] = False 
plt.plot(x,ft(x))

plt.ylabel("传感器的容量(mA/h)")
plt.xlabel("充电器休息时间(h)")
plt.grid()
plt.savefig('ft.png')
plt.show()

```

```python
from mpl_toolkits import mplot3d
ax = plt.axes(projection='3d')
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
#三维线的数据
x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)
X, Y = np.meshgrid(x, y)
Z = f(X,Y)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
ax.set_title('surface')
```
