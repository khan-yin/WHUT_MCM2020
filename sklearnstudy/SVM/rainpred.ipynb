```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
weather=pd.read_csv("weatherAUS.csv")
```

```python
weather.info()
```

```python
weather.head()
```

```python
weather=weather.iloc[:5000,:]
```

```python
weather.shape
```

```python
X=weather.iloc[:,:-1]
```

```python
Y=weather.iloc[:,-1]
```

```python
X.info()
```

```python
#探索缺失值
X.isnull().mean()#均值表示缺失值所占比例
```

```python
#探索标签的分类
np.unique(Y)
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.3,random_state=420)
```

```python
Xtrain.head()
```

```python
#恢复索引
for i in [Xtrain, Xtest, Ytrain, Ytest]:
    i.index = range(i.shape[0])
```

```python
Ytrain.head()
```

```python
#是否有样本不平衡问题？
Ytrain.value_counts()
#有轻微的不均衡
```

```python
Ytrain.value_counts()[0]/Ytrain.value_counts()[1]
```

```python
Ytest.value_counts()
```

```python
#将标签编码
from sklearn.preprocessing import LabelEncoder
encorder = LabelEncoder().fit(Ytrain)
Ytrain = pd.DataFrame(encorder.transform(Ytrain))
Ytest = pd.DataFrame(encorder.transform(Ytest))
```
