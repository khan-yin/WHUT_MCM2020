```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(color_codes=True)
```

```python
np.random.seed(sum(map(ord,"distributions")))
```

```python
x=np.random.normal(size=100)
sns.distplot(x,kde=False)
```

```python
sns.distplot(x,bins=10,kde=False)
```

```python
x=np.random.gamma(6,size=200)
```

```python
from scipy import stats,integrate
#scipy.stats统计分布
sns.distplot(x,kde=False,fit=stats.gamma)#gama分布
```

```python
mean,cov=[0,1],[(1,.5),(.5,1)]
```

```python
data=np.random.multivariate_normal(mean,cov,200)
```

```python
df=pd.DataFrame(data,columns=["x","y"])
```

```python
df.head()
```

```python
import scipy.stats as sci
sns.jointplot(x="x",y="y",data=df,kind="scatter",color='b',stat_func=sci.pearsonr)
```

```python
x,y=np.random.multivariate_normal(mean,cov,10000).T
```

```python
with sns.axes_style("white"):
    sns.jointplot(x=x,y=y,kind="hex",color="k",stat_func=sci.pearsonr)
```
