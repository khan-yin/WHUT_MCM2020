### 2. 绘制各省会城市空气质量情况

历史天气：http://tianqi.2345.com/wea_history/59287.htm

```python
import pandas as pd
```

```python
qg = pd.read_excel('./data/data.xlsx',sheetname='全国')
```

```python
qg.head(2)
```

```python
air = qg['空气质量指数'].tolist()
print(min(air),max(air))
```

```python
city = qg['城市'].tolist()
```

- 配色网站：https://color.adobe.com/zh/explore/newest/

```python
from pyecharts import Geo,Style

# 图形整体配置文件
style = Style(
    title_color="#fff",
    title_pos = "center",
    width=980,
    height=600,
    background_color="#404a59"
)

# 图形局部细节配置文件
style_geo = style.add(
    type="effectScatter",
    is_visualmap=True,          # 开启数值到颜色或形状的视觉映射
    visual_range=[0, 350],
    visual_text_color="#fff",    # 设置视觉组件的文本颜色
    symbol_size=15,
    border_color="#fff",
    geo_normal_color="#323c48",
    geo_emphasis_color="#2a333d",
    geo_cities_coords=None,
    is_roam=True
)

geo = Geo("各省会城市空气质量",**style.init_style)
geo.add("",city,air,**style_geo)
geo.render()
geo
```

### 3.绘制省份内不同城市的公众号粉丝数分布情况

```python
sf = pd.read_excel('./data/data.xlsx',sheetname='省份')
```

```python
sf
```

```python
city = sf['城市'].tolist()
nums = sf['用户数'].tolist()
```

```python
print(min(nums),max(nums))
```

```python
from pyecharts import Geo,Style

# 图形整体配置文件
style = Style(
    title_color="#fff",
    title_pos = "center",
    width=980,
    height=600,
    background_color="#404a59"
)

# 图形局部细节配置文件
style_geo = style.add(
    type="effectScatter",
    maptype="湖北",
    is_visualmap=True,
    visual_range=[0, 450],
    visual_text_color="#fff"
)


geo = Geo("某公众号在湖北省不同城市的粉丝数",**style.init_style)
geo.add("",city,nums,**style_geo)
geo.render()
geo
```

### 4. 用自定义的城市坐标覆盖原来的城市坐标

- 高德地图坐标拾取器：https://lbs.amap.com/console/show/picker

```python
city_geo = pd.read_excel("./data/data.xlsx",sheetname='city_geo')
city_geo.head(2)
```

```python
city_geo.index = city_geo['city']
city_geo = city_geo.drop('city',axis=1)
city_geo.head(2)
```

```python
city_geo = city_geo.T.to_dict(orient='list')
```

```python
city_geo
```

```python
from pyecharts import Geo,Style

# 图形整体配置文件
style = Style(
    title_color="#fff",
    title_pos = "center",
    width=980,
    height=600,
    background_color="#404a59"
)

# 图形局部细节配置文件
style_geo = style.add(
    type="effectScatter",
    maptype="湖北",
    is_visualmap=True,
    visual_range=[0, 450],
    visual_text_color="#fff",
    geo_cities_coords=city_geo
)


geo = Geo("某公众号在湖北省不同城市的粉丝数",**style.init_style)
geo.add("",city,nums,**style_geo)
geo.render()
geo
```
