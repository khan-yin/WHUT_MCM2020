### 2.对真实的数据进行可视化

```python
import pandas as pd
user_analysis = pd.read_excel("./data/user_analysis.xlsx")
user_analysis.head(2)
```

- 此数据集记录的是2015年~2018年的某公众号的粉丝增长数据；
- 希望对比这三年每月的粉丝净增长情况；
- 需根据提供的原始数据进行汇总和处理，得到分析结论；

```python
user_analysis['年份'] = user_analysis['时间'].apply(lambda s:s.year)
user_analysis['月份'] = user_analysis['时间'].apply(lambda s:s.month)
```

```python
user_analysis.head(2)
```

```python
user_data = user_analysis.groupby(['月份','年份'],as_index=False)['净增关注人数'].sum()
user_data.head()
```

```python
user_data['年份'] = user_data['年份']-2015
user_data['月份'] = user_data['月份']-1
user_data.head(10)
```

```python
from pyecharts import Bar3D,Style

style = Style(
    title_color="#000",
    title_pos = "center",
    width=980,
    height=600,
    background_color="#fff"
)

style_3d = style.add(
    is_visualmap=True,
    visual_range=[-200,1000],
    visual_range_color=['#313695', '#4575b4', '#74add1', '#abd9e9', '#e0f3f8', '#ffffbf',
                        '#fee090', '#fdae61', '#f46d43', '#d73027', '#a50026'],
    grid3d_width=200, 
    grid3d_depth=80,
    xaxis_label_textcolor='#fff',
    is_grid3d_rotate=True,
    legend_pos='left'
)

x_axis = ['1月','2月','3月','4月','5月','6月','7月','8月','9月','10月','11月','12月']
y_axis = ['2015年','2016年','2017年','2018年']

bar3d = Bar3D("3年来公众号粉丝增长数据可视化",**style.init_style)
bar3d.add("每月粉丝净增长",x_axis,y_axis,user_data.values.tolist(),**style_3d)
bar3d.render()
bar3d
```
