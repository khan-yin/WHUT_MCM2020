### 1. 导入需要使用的库

```python
import pandas as pd
from pyecharts import GeoLines, Style
```

### 2.数据预处理

#### 火车路线数据获取

- 经过某个城市的火车线路数据可以通过12306火车票订票网站获取，地址：https://kyfw.12306.cn/otn/czxx/init
- 数据在Excel中进行适当的预处理；

```python
trains = pd.read_excel('./data/铁路数据.xlsx')
trains
```

```python
bj = pd.read_excel("./data/铁路数据-北京.xlsx")
bj.head()
```

```python
trains = trains[['始发站','终到站']]
bj = bj[['始发站','终到站']]
```

```python
def dropDNXB(s):
    except_citys = set(['济南','靖西'])
    
    dnxb = set(["东","南","西","北"])
    
    if except_citys.issuperset(set([s])):
        return s
    elif dnxb.issuperset(s[-1]):
        return s[:-1]
    else:
        return s
```

```python
dropDNXB('济南')
```

```python
dropDNXB('南宁东')
```

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
trains.loc[:,'始发站'] = trains['始发站'].apply(lambda s:dropDNXB(s))
trains.loc[:,'终到站'] = trains['终到站'].apply(lambda s:dropDNXB(s))
```

```python
bj.loc[:,'始发站'] = bj['始发站'].apply(lambda s:dropDNXB(s))
bj.loc[:,'终到站'] = bj['终到站'].apply(lambda s:dropDNXB(s))
```

```python
trains = trains.drop_duplicates()
```

```python
bj = bj.drop_duplicates()
```

```python
bj_pairs = bj.values.tolist()
```

```python
trains_pairs = trains.values.tolist()
trains_pairs
```

### 3.可视化展示

```python
style = Style(
    title_color="#fff",
    title_pos = "center",
    width=980,
    height=600,
    background_color="#404a59"
)

style_geo = style.add(
    maptype="china",
    is_label_show=False,
    line_curve=0.2,
    line_opacity=0.6,
    legend_text_color="#eee",
    legend_pos="right",
    geo_effect_symbol="triangle",
    geo_effect_symbolsize=6,
    symbol_size=8,
    label_color=['#a6c84c', '#ffa022', '#46bee9'],
    label_pos="right",
    label_formatter="{b}",
    label_text_color="#eee",
    is_legend_show=True
)

geolines = GeoLines("铁路线路", **style.init_style)
geolines.add(name = "经过南宁",data = trains_pairs,**style_geo)
geolines.add(name = "经过北京",data = bj_pairs,**style_geo)
geolines.render('./html/trains.html')
geolines
```
